### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
missing_values = df.isnull().sum()
print("=== Missing Values per Column ===")
print(missing_values)
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\n=== Percentage of Missing Values per Column ===")
print(missing_percentage)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
## Write your code here

# Task 3: Visualize missing values
# Heatmap (only columns that actually have NaNs)
plt.figure(figsize=(12, 6))
cols_with_na = df.columns[df.isna().any()]
sns.heatmap(df[cols_with_na].isna(), cbar=False, cmap="YlGnBu", yticklabels=False)
plt.title("Missing Values Heatmap (columns with NaNs only)")
plt.show()

# Bar chart: % missing per column (sorted, non-zero only)
missing_pct = (df.isna().mean() * 100).loc[lambda s: s > 0].sort_values(ascending=False)
plt.figure(figsize=(12, 6))
missing_pct.plot(kind="bar", color="salmon")
plt.title("Percentage of Missing Values per Column")
plt.ylabel("Percent (%)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here

# Task 4: Missing count for 'Employment'
col = "Employment"
miss_n = df[col].isna().sum()
miss_pct = df[col].isna().mean() * 100
print(f"Missing '{col}': {miss_n} rows ({miss_pct:.2f}%)")
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
# Task 5: Most frequent (majority) value in 'Employment'
mode_vals = df[col].mode(dropna=True)
most_frequent = mode_vals.iloc[0] if not mode_vals.empty else None
print(f"Most frequent '{col}' value:", most_frequent)
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here

# Task 6: Imputation
# (Optional) keep a copy for comparison in Task 7
df["Employment_before"] = df[col].copy()

df[col] = df[col].fillna(most_frequent)

# Verify
print(f"Missing '{col}' after imputation:", df[col].isna().sum())
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
# Task 7: Distribution AFTER imputation (clean view = top 15, horizontal)
plt.figure(figsize=(10, 8))
df[col].value_counts().head(15).plot(kind="barh", color="steelblue")
plt.title("Top 15 'Employment' Categories After Imputation")
plt.xlabel("Count"); plt.ylabel("Employment")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Compare side by side
# %%
fig, axes = plt.subplots(1, 2, figsize=(16, 8), sharex=False)

# Before (top 10)
df["Employment_before"].value_counts().head(10).plot(kind="barh", ax=axes[0], color="darkorange")
axes[0].set_title("Employment (Before Imputation)")
axes[0].set_xlabel("Count"); axes[0].invert_yaxis()

# After (top 10)
df[col].value_counts().head(10).plot(kind="barh", ax=axes[1], color="steelblue")
axes[1].set_title("Employment (After Imputation)")
axes[1].set_xlabel("Count"); axes[1].invert_yaxis()

plt.tight_layout()
plt.show()
```
